#### Метрики ефективності email-комунікації для покупців

Для оцінки ефективності email-комунікації з покупцями я розрахувала базові метрики взаємодії: кількість надісланих повідомлень, відкриттів, кліків та конверсій. На основі цих значень також було обчислено показники open rate, click-through rate (CTR), click-to-open rate (CTOR) та conversion rate. Окремо метрики були розраховані для першого повідомлення та повторних повідомлень.


```python
#Створила окрему вибірку із покупцями, для того щоб краще їх бачити
buyer = df[df['buyer'] == 'Buyer'].copy()
```

```python
buyer.head(2)
```

```python
buyer['response'].value_counts()
```

hypotesis
Перша email-відправка з повідомленням не створює достатнього тригера для повернення платника до діалогу, що призводить до необхідності повторних нагадувань.

```python
#Домінування повторних message-івентів вказує на можливу неефективність першої спроби повернення користувача до діалогу
plt.pie(buyer['response'].value_counts(), labels=buyer['response'].unique(),autopct='%1.1f%%')
plt.show()
```

```python
first_msg = buyer[buyer['response'] == 'unread_unused_message']
repeat_msg = buyer[buyer['response'] == 'unread_used_message']
```

```python
def email_metrics(df_part):
    sent = len(df_part)
    opened = df_part['read_ts'].notna().sum()
    clicked = df_part['click_ts'].notna().sum()
    converted = (df_part['total_credits'] > 0).sum()

    return {
        'sent': sent,
        'opened': opened,
        'clicked': clicked,
        'open_rate': opened / sent if sent > 0 else 0,
        'ctr': clicked / sent if sent > 0 else 0,
        'ctor': clicked / opened if opened > 0 else 0,
        'conversion_rate': converted / sent if sent > 0 else 0
    }
```

```python
first_metrics = email_metrics(first_msg)
repeat_metrics = email_metrics(repeat_msg)
```

```python
pd.DataFrame({
    "Метрика": first_metrics.keys(),
    "Перше повідомлення": first_metrics.values(),
    "Повторне повідомлення": repeat_metrics.values()
})
```

#### Оцінка сегментної логіки email-розсилок


```python
rule_stats = (
    buyer
    .groupby('rule')
    .agg(
        sent=('rule', 'count'),
        clicked=('click_ts', lambda x: x.notna().sum())
    )
    .reset_index()
)

rule_stats['ctr'] = rule_stats['clicked'] / rule_stats['sent']
```

```python
rule_stats.sort_values('ctr', ascending=False)
```

#### Потрібний час для взаємодії після email-повідомлення
У цій підсекції я проаналізувала час, необхідний покупцям для здійснення дії після отримання email-повідомлення. Для цього було відібрано лише ті випадки, де користувач здійснив транзакцію (витратив кредити), та розраховано час між моментом надсилання листа і кліком, виражений у годинах.


```python
buyer['time_to_open_hours'] = (buyer['read_ts'] - buyer['send_ts']) / 3600
```

```python
tto = buyer[buyer['time_to_open_hours'].notna()]
```

```python
#Час до відкриття листа
tto['time_to_open_hours'].describe()
```

```python
buyer['time_to_click_hours'] = (buyer['click_ts'] - buyer['send_ts']) / 3600
ttc = buyer[buyer['time_to_click_hours'].notna()]
```

```python
#Час до кліку
ttc['time_to_click_hours'].describe()
```

```python
buyers_conv = buyer[buyer['total_credits'] > 0].copy()
buyers_conv['time_to_conversion_hours'] = (buyers_conv['click_ts'] - buyers_conv['send_ts']) / 3600
```

```python
#Час до конверсії
buyers_conv['time_to_conversion_hours'].describe()
```
